```python
import numpy as np
import matplotlib.pyplot as plt
#import pandas as pd
```

```python
name='digue_1D_an'
data = np.loadtxt('DC_digue1D_an.txt',skiprows = 1)
pick = np.loadtxt('save.txt')
mode0 = np.loadtxt('120mode0.txt')
mode1 = np.loadtxt('120mode1.txt')
mode2 = np.loadtxt('120mode2.txt')
mode3 = np.loadtxt('120mode3.txt')



mail_fmax=150
```

```python
plt.figure(figsize= (15,8))
plt.plot(mode0[:,0],1/mode0[:,1],'midnightblue', label="mode0")
plt.plot(mode1[:,0],1/mode1[:,1],'blue', label="mode1")
plt.plot(mode2[:,0],1/mode2[:,1],'royalblue', label="mode2")
plt.plot(mode3[:,0],1/mode3[:,1],'lightsteelblue', label="mode3")
plt.grid()
plt.legend()
plt.xlabel('Frequence (Hz)')
plt.ylabel('Vitesse de phase (m/s)')
plt.tick_params(axis = 'both', labelsize =15)

figtheo="_theorical.png"
namefigtheo=name+figtheo
#plt.savefig(namefigtheo,bbox_inches='tight')
```

```python
x = np.reshape(data[:,0],(-1,len(np.unique(data[:,0]))))
y = 1/(np.reshape(data[:,1],(-1,len(np.unique(data[:,0])))))
z1 = np.reshape(data[:,2],(-1,len(np.unique(data[:,0]))))
```

```python

#courbes limites
L=24
x1=5
dx=1

wlmax=0.5*L
print('wlmax4 (edf : wl< 0.5L)=', wlmax, 'm')


wlmin=2*dx
print('wlmin (pour aliasing quand wl < 2*dx)=', wlmin, "m")


#plot
plt.figure(figsize=(20,10))
cf = plt.contourf(x,y,-z1,20, cmap='RdYlBu')
plt.colorbar(label = 'val')
plt.contour(cf,colors = 'k',linewidths = 0.3,alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')

v=np.arange(50,800,10)         
fmin=v/wlmax
fmax=v/wlmin

plt.plot(fmin,v, 'k--', label=" wl< 0.5L")

plt.plot(fmax,v,'k', label="Aliasing quand wl < 2*dx")

#plt.plot(pick[0,:],pick[1,:],'k', label="DC terrain")

plt.xlim(10,150)
plt.ylim(50,500)
plt.xlabel('Frequence (Hz)')
plt.ylabel('Vitesse de phase (m/s)')

plt.plot([mail_fmax, mail_fmax],[50,800],  'w--', lw=2, label="Frquence limite du maillage")

plt.tick_params(axis = 'both', labelsize = 15)

#plt.plot([0, 1500],[157.5,157.5],  'w--', lw=2, label="160 m/s")
#plt.plot([0, 1500],[324,324],  'w', lw=2, label="320 m/s")
plt.plot(mode0[:,0],1/mode0[:,1], 'w', label="Mode 0")
#plt.plot(mode1[:,0],1/mode1[:,1], 'wheat', label="Mode 1")
#plt.plot(mode2[:,0],1/mode2[:,1], 'sandybrown', label="Mode 2")
#plt.plot(mode3[:,0],1/mode3[:,1], 'darkorange',label="Mode 3")


plt.legend()
plt.grid()

figfil="_compa_tr_1D.png"
namefigfil=name+figfil
plt.savefig(namefigfil,bbox_inches='tight')
```
