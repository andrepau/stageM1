```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft 
from numpy.fft import ifft 
import scipy as sp
from scipy.signal import butter, lfilter, filtfilt
from scipy.signal import freqz
```

```python
digue1D = np.transpose(np.loadtxt('pick_digue1D.txt'))
digue1D_2C = np.transpose(np.loadtxt('pick_digue1D_2C.txt'))
digue2D = np.transpose(np.loadtxt('pick_digue2D.txt'))
digue2D_2C = np.transpose(np.loadtxt('pick_digue2D_2C.txt'))
digue3D = np.transpose(np.loadtxt('pick_digue3D.txt'))
digue3D_2C = np.transpose(np.loadtxt('pick_digue3D_2C.txt'))

digue1Dan = np.transpose(np.loadtxt('pick_digue1Dan.txt'))
digue1D_2Can = np.transpose(np.loadtxt('pick_digue1D_2Can.txt'))
digue3Dan = np.transpose(np.loadtxt('pick_digue3Dan.txt'))
digue3D_2Can = np.transpose(np.loadtxt('pick_digue3D_2Can.txt'))

terrain=np.transpose(np.loadtxt('pick_P0_02.txt'))
```

```python
np.shape(digue3D)
```

```python
#####
name='test'
#####


plt.figure(figsize= (15,8))

plt.plot(digue1D[:,0],digue1D[:,1],'orange', label='digue1D')
plt.plot(digue2D[:,0],digue2D[:,1],'g', label='digue2D')
plt.plot(digue3D[:,0],digue3D[:,1], label='digue3D')
plt.plot(terrain[:,0],terrain[:,1], '--r', label='Data terrain')

plt.grid()
plt.legend()
plt.xlabel('Frequence (Hz)')
plt.ylabel('Vitesse de phase (m/s)')
plt.tick_params(axis = 'both', labelsize =15)
plt.xlim(10,150)

fig="_pickcomp.png"
namefig=name+fig
plt.savefig(namefig,bbox_inches='tight')
```
