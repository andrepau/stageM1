```python
#programme .py pour calculer et creer le fichier .fsr
#position des recepteurs

import numpy as np
```

```python
xsource=input('x source:')
xsource=int(xsource)

ysource=input('y source:')
ysource=int(ysource)

x1=input('distance source 1er recepteur:')
x1=int(x1)

nb=input('nb de recepteurs:')
nb=int(nb)

dx=input('distance entre recepteurs:')
dx=int(dx)


xrec=np.arange((xsource+x1),(xsource+x1)+((nb)*dx),dx)
yrec=np.linspace(ysource,ysource,nb)
xrec=xrec.astype(int)
yrec=yrec.astype(int)

xy=[xrec,yrec]
xy=np.transpose(xy)

print(xy)
#print(xy,np.size(xrec),np.size(yrec))
np.savetxt("digue3.fsr",(xy), fmt='%10.5f' )
```

```python
xsource=input('x source:')
xsource=float(xsource)

ysource=input('y source:')
ysource=float(ysource)

x1=input('distance source 1er recepteur:')
x1=float(x1)

nb=input('nb de recepteurs:')
nb=int(nb)

dx=input('distance entre recepteurs:')
dx=float(dx)

xrec=np.arange((xsource+x1),(xsource+x1)+((nb)*dx),dx)
yrec=np.linspace(ysource,ysource,nb)

R=["..."]*nb
for i in range(nb):
    if i < 9:
        num=str(i+1)
        name='R00'
        name=name+num
    else:
        num=str(i+1)
        name='R0'
        name=name+num
    R[i]=name

R=np.transpose(R)
xy=[R,xrec,yrec]
xy=np.transpose(xy)

print(xy)
print(type(xy))
#print(xy,np.size(xrec),np.size(yrec))

```

```python
np.savetxt("gao3.coord",xy,fmt="%s" )
```
