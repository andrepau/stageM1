```python
import math
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft 
from numpy.fft import ifft 
```

```python
data= np.loadtxt('gao1.sff',skiprows=1)
t=data[:,0]
a=data[:,1]

plt.figure(figsize=(20,10))

plt.plot(t,a)
plt.grid()
```

```python
print(t)
tfd = fft(a)
N=len(a)
spectre = np.absolute(tfd)*2/N
T=max(t)

freq=np.arange(N)*1.0/T
#freq=1/t 
print(freq)

plt.figure(figsize=(10,4))
plt.plot(freq,spectre,'r')
plt.xlabel('f')
plt.ylabel('A')
plt.xlim(-10,80)
plt.grid()
```

```python
def plot_fft(signal,t):
    plt.figure(figsize=(20,10))
    plt.subplot(211)
    plt.plot(t,signal)
    plt.grid()
    plt.xlabel('temps (s)')
    plt.ylabel('Amplitude')
    
    tfd = fft(signal)
    N=len(signal)
    spectre = np.absolute(tfd)*2/N
    T=max(t)

    freq=np.arange(N)*1.0/T
    

    plt.subplot(212)
    plt.plot(freq,spectre,'r')
    plt.xlabel('fréquence (Hz)')
    plt.ylabel('Amplitude')
    plt.xlim(-5,100)
    plt.grid()
```

```python
plot_fft(a,t)
```
